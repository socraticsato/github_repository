```python
import json

# Example data (Python dictionary)
data = {
    "name": "Sena Sato",
    "age": 9,
    "sports": ["tennis", "soccer"],
    "ranking": {"national": 18, "local": 7}
}

# --- Write JSON data to a file ---
with open("player.json", "w") as file:
    json.dump(data, file, indent=4)  # indent=4 makes it readable

print("✅ JSON file created successfully!")

# --- Read JSON data from the file ---
with open("player.json", "r") as file:
    loaded_data = json.load(file)

print("\n📖 Data loaded from file:")
print(loaded_data)

# --- Update the JSON data ---
loaded_data["ranking"]["national"] = 25  # Update a value
loaded_data["sports"].append("basketball")  # Add a new item

# Save changes back to the file
with open("player.json", "w") as file:
    json.dump(loaded_data, file, indent=4)

print("\n✅ JSON file updated successfully!")

```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# Sample training data
texts = ["Buy cheap products now", "Limited offer just for you", "Meeting at 10am", "Project deadline tomorrow"]
labels = ["spam", "spam", "ham", "ham"]  # ham = not spam

# Convert text to numeric features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Predict
test = vectorizer.transform(["Luxurious deal just for you."])
print(model.predict(test))  # Output: ['spam']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# Sample training data
texts = ["Buy cheap products now", "Limited offer just for you", "Meeting at 10am", "Project deadline tomorrow"]
labels = ["spam", "spam", "ham", "ham"]  # ham = not spam

# Convert text to numeric features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Predict
test = vectorizer.transform(["cheap offer just for you"])
print(model.predict(test))  # Output: ['spam']
```

```python
# 🧩 Install dependencies first (uncomment if running for the first time)
# !pip install transformers torch

from transformers import AutoTokenizer, AutoModel
import torch

# Step 1: Choose a pretrained model (BERT-base for this example)
model_name = "bert-base-uncased"

# Step 2: Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Step 3: Define input text
text = "Japan must embrace change to thrive in the age of AI."

# Step 4: Tokenize text
inputs = tokenizer(text, return_tensors="pt")
print("🔹 Tokens:", tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]))
print("🔹 Token IDs:", inputs["input_ids"])

# Step 5: Get embeddings from model
with torch.no_grad():
    outputs = model(**inputs)

# The last hidden state is the embedding tensor for each token
embeddings = outputs.last_hidden_state

print("🔹 Embedding shape:", embeddings.shape)
print("   (batch_size, num_tokens, embedding_dimension)")

# Step 6: Example — average token embeddings to get sentence embedding
sentence_embedding = embeddings.mean(dim=1)
print("🔹 Sentence embedding shape:", sentence_embedding.shape)
print("🔹 Sentence embedding vector (first 5 values):")
print(sentence_embedding) # Completed the print statement
```
